## working with dates

```python
# converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

### Extract year

```python
date['date_year']=date['date'].dt.year
date.head()
```

### Extract month

```python
date['date_month']=date['date'].dt.month
date.head()
```

### Extract month name

```python
date['date_month_name']=date['date'].dt.month_name()
date.head()
```

### Extract day

```python
date['date_days']=date['date'].dt.day
date.head()
```

### Extract Day of week

```python
date['day_of_week']=date['date'].dt.dayofweek
date.head()
```

### Extract Day name

```python
date['day_name']=date['date'].dt.day_name()
date.head()
```

### check is weekend?

```python
date['date_is_weekend']=np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract week of year

#### particular date is resides in which week in total weeks it return week number

```python
date["date_week"]=date['date'].dt.week
date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract quarter

```python
date["date_quarter"]=date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).sample(5)
```

### Extract semester

```python
date['semester']=np.where(date['date_quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).sample(5)
```

### Extract Time elapsed between dates

```python
import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
# day passed
(today-date['date']).dt.days
```

```python
## month passed
np.round((today-date['date'])/np.timedelta64(1,'M'),0)
```
