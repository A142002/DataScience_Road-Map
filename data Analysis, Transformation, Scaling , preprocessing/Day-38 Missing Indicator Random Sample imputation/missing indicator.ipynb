```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df=pd.read_csv("train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
si=SimpleImputer()  #by default all missing values replace by mean
x_train_trf=si.fit_transform(x_train)
x_test_trf=si.transform(x_test)
```

```python
x_train_trf
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train_trf,y_train)
y_pred=clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
mi=MissingIndicator()
mi.fit(x_train)
```

```python
mi.features_   #indicate index 0 column contain missing value
```

```python
x_train_missing=mi.transform(x_train)
x_test_missing=mi.transform(x_test)
```

```python
x_train_missing
```

```python
x_test_missing
```

```python
x_train['Age_NA']=x_train_missing
x_test['Age_NA']=x_test_missing
```

```python
x_train.sample(5)
```

```python
si=SimpleImputer()  #by default all missing values replace by mean
x_train_trf2=si.fit_transform(x_train)
x_test_trf2=si.transform(x_test)
```

```python
## Accuracy get increase after missing indicator

from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train_trf2,y_train)
y_pred=clf.predict(x_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Here i use add indicator it directly add missing indicator
#no need to use MissingIndicator class
si=SimpleImputer(add_indicator=True)   
x_train_trf3=si.fit_transform(x_train)
x_test_trf3=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train_trf3,y_train)
y_pred=clf.predict(x_test_trf3)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
